# Feature Extraction

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import csv
from tqdm import tqdm_notebook

def get_image_features(img, x1, y1, x2, y2, verbose=False):
    img = img[x1:x2, y1:y2]
    dst = cv2.fastNlMeansDenoising(img)
    noise = img - dst
    resized_noise = cv2.resize(noise, (20, 20))
    if verbose:
        print(np.mean(noise), np.std(noise))
        fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20, 4))
        ax1.imshow(img, cmap='gray')
        ax2.imshow(dst, cmap='gray')
        ax3.imshow(noise, cmap='gray')
        ax4.imshow(resized_noise, cmap='gray')
    return np.concatenate(([np.mean(noise), np.std(noise)], resized_noise.flatten()))


def get_path_features(path):
    features_list = []
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    for rot in [0, 90, 90, 90]:
        if rot != 0:
            num_rows, num_cols = img.shape[:2]
            rotation_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2), rot, 1)
            img = cv2.warpAffine(img, rotation_matrix, (num_cols, num_rows))
        for _ in range(0, 10):
            x, y = np.random.randint(img.shape[0]-512), np.random.randint(img.shape[1]-512)
            features_list.append(get_image_features(img, x, y, x + 512, y + 512))
    return features_list

def save_all_features(paths, output):
    with open(output, 'w') as myfile:
        wr = csv.writer(myfile, quoting=csv.QUOTE_NONNUMERIC)
        for path in tqdm_notebook(paths):
            features_list = get_path_features(path)
            for features in features_list:
                wr.writerow(features.tolist() + [os.path.basename(os.path.dirname(path)), os.path.basename(path)])

paths = ['train/Samsung-Galaxy-Note3/(GalaxyN3)54.jpg']
save_all_features(paths, 'features.csv')
```

```python
import os

paths = []
base_dir = '/media/carlosbaia/HD - CARLOS HENRIQUE/Unicamp/MO444/train'
for device_dir in os.listdir(base_dir):
    for file in os.listdir(os.path.join(base_dir, device_dir)):
        paths.append(os.path.join(base_dir, device_dir, file))
      
print('Extracting features of {} images...'.format(len(paths)))
save_all_features(paths, 'features.csv')
```
